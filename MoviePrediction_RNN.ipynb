```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow .keras.preprocessing import sequence
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,SimpleRNN,Dense
```

```python
max_features=10000
(X_train,y_train),(X_test,y_test)=imdb.load_data(num_words=max_features)
print("Training data shape",X_train.shape,"Training labels shape",y_train.shape)
print("Training data shape",X_test.shape,"Training labels shape",y_test.shape)
```

```python
sample_review=X_train[0]
sample_label=y_train[0]
print("Review",sample_review)
print("Label",sample_label)
```

```python
word_index=imdb.get_word_index()
reverse_word_index={value:key for key , value in word_index.items()}
reverse_word_index
```

```python
decoded_review=' '.join([reverse_word_index.get(i -3,'?')for i in sample_review])
decoded_review
```

```python
from tensorflow.keras.preprocessing import sequence
max_len=500
X_train=sequence.pad_sequences(X_train,maxlen=max_len)
X_test=sequence.pad_sequences(X_test,maxlen=max_len)
X_train
```

```python
X_train[0]
```

```python
from keras.models import Sequential
from keras.layers import Embedding, SimpleRNN, Dense

model = Sequential()
model.add(Embedding(input_dim=max_features, output_dim=128, input_length=max_len))
model.add(SimpleRNN(128, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.build(input_shape=(None, max_len))
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping
earlystopping=EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)
earlystopping
```

```python
history=model.fit(X_train,y_train,epochs=10,batch_size=32,validation_split=0.2,
                  callbacks=[earlystopping])
```

```python
model.save('movie_review_rnn.h5')
```

```python
from tensorflow.keras.models import load_model
model=load_model('movie_review_rnn.h5')
model.summary()
```

```python
model.get_weights()
```

```python
def decode_review(encoded_review):
  return ' '.join([reverse_word_index.get(i -3,'?')for i in encoded_review])
def preprocess_text(text):
  words=text.lower().split()
  encoded_review=[word_index.get(word,2)+3 for word in words]
  padded_review=sequence.pad_sequences([encoded_review],maxlen=500)
  return padded_review
```

```python
def predict_sentiment(review):
  preprocessed_input=preprocess_text(review)
  prediction=model.predict(preprocessed_input)
  sentiment='Positive' if prediction[0][0]>0.5 else 'Negative'
  return sentiment , prediction[0][0]
```

```python
example="This movie was fantastic"
sentiment,score=predict_sentiment(example)
print("Review",example)
print("Sentiment",sentiment)
print("Prediction Score",score)
```
